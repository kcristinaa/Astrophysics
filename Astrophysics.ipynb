```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))

font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 20,
        }

with open('profile.out','r') as f:
    lines = f.readlines()
    x = [float(line.split()[0]) for line in lines]
    y = [float(line.split()[4]) for line in lines]
    
with open('profile1.out','r') as f:
    lines = f.readlines()
    a = [float(line.split()[0]) for line in lines]
    b = [float(line.split()[4]) for line in lines]
    
with open('profile20.out','r') as f:
    lines = f.readlines()
    c = [float(line.split()[0]) for line in lines]
    d = [float(line.split()[4]) for line in lines]
    
with open('nr1.out','r') as f:
    lines = f.readlines()
    e = [float(line.split()[0]) for line in lines]
    g = [float(line.split()[4]) for line in lines]
    
with open('nr2.5.out','r') as f:
    lines = f.readlines()
    h = [float(line.split()[0]) for line in lines]
    i = [float(line.split()[4]) for line in lines]
    
with open('nr20.out','r') as f:
    lines = f.readlines()
    j = [float(line.split()[0]) for line in lines]
    k = [float(line.split()[4]) for line in lines]
    
fig1 = plt.gcf()   
plt.plot(c,d,'g',label="x$_{FC}$ = 20, relativistic")
plt.plot(j,k,'m',label="x$_{FC}$ = 20, non-relativistic")
plt.plot(x,y,'r', label="x$_{FC}$ = 2.5, relativistic")
plt.plot(h,i,'k',label="x$_{FC}$ = 2.5, non-relativistic")
plt.plot(a,b,'b',label="x$_{FC}$ = 1, relativistic")
plt.plot(e,g,'y',label="x$_{FC}$ = 1, non-relativistic")
plt.ylabel('Density (kg/m$^{3}$)',fontweight="bold", fontdict=font)
plt.xlabel('Distance r(km)',fontweight="bold", fontdict=font)
plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20) 
plt.yscale("log")
plt.ylim([10**(-7), 10**4])
plt.xlim([0,10000])
plt.grid(b=None, which='major', axis='both')
plt.legend(prop={'size': 25})
plt.show()

plt.draw()
fig1.savefig('profilecompare.jpg', dpi=100)
```
